```python
import requests
from bs4 import BeautifulSoup

import pandas as pd
```

```python
headers = {'User-Agent': 
           'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36'}
```

```python
def getTransfers(year):
    page = "https://www.transfermarkt.co.uk/transfers/transferrekorde/statistik?saison_id="+year+"&land_id=0&ausrichtung=&spielerposition_id=&altersklasse=&leihe=&w_s=+1&plus=1"
    pageTree = requests.get(page, headers=headers)
    pageSoup = BeautifulSoup(pageTree.content, 'html.parser')
    
    Players = pageSoup.find_all("a", {"class": "spielprofil_tooltip"})
    Values = pageSoup.find_all("td", {"class": "rechts hauptlink"})
    Ages = pageSoup.find_all("td", {"class": "zentriert"})[1::4]
    Left = pageSoup.find_all("a", {"class": "vereinprofil_tooltip"})[1::4]
    Joined = pageSoup.find_all("a", {"class": "vereinprofil_tooltip"})[3::4]
    PlayersList = []
    ValuesList = []
    AgesList = []
    LeftList = []
    JoinedList = [] 

    for i in range(0,25):
        PlayersList.append(Players[i].text)
        ValuesList.append(Values[i].text)
        AgesList.append(Ages[i].text)
        LeftList.append(Left[i].text)
        JoinedList.append(Joined[i].text)
        df = pd.DataFrame({"Player":PlayersList,"Fee":ValuesList,"Left":LeftList,"Joined":JoinedList,"Ages":AgesList})
        
    return df
```

```python
years = ['2012','2013','2014','2015','2016','2017','2018']
transfers_temp = {}
for year in years:
    transfers_temp[year] = getTransfers(year)
    transfers_temp[year]["Year"] = year
```

```python
transfers = transfers_temp['2012'].append(transfers_temp['2013'], ignore_index = True)
transfers = transfers.append(transfers_temp['2014'], ignore_index=True)
transfers = transfers.append(transfers_temp['2015'], ignore_index=True)
transfers = transfers.append(transfers_temp['2016'], ignore_index=True)
transfers = transfers.append(transfers_temp['2017'], ignore_index=True)
transfers = transfers.append(transfers_temp['2018'], ignore_index=True)

transfers['Fee'] = [str(s)[1:-1] for s in transfers['Fee']]
```

```python
transfers['Player'] = transfers['Player'].str.replace('á','a')
transfers['Player'] = transfers['Player'].str.replace('í','i')
transfers['Player'] = transfers['Player'].str.replace('Á','A')
transfers['Player'] = transfers['Player'].str.replace('é','e')
transfers['Player'] = transfers['Player'].str.replace('à','a')
transfers['Player'] = transfers['Player'].str.replace('ú','u')
transfers['Player'] = transfers['Player'].str.replace('ç','c')
transfers['Player'] = transfers['Player'].str.replace('ã','a')
transfers['Player'] = transfers['Player'].str.replace('ü','u')
transfers['Player'] = transfers['Player'].str.replace('ï','i')

transfers['Left'] = transfers['Left'].str.replace('é','e')
transfers['Left'] = transfers['Left'].str.replace('ã','a')
transfers['Joined'] = transfers['Joined'].str.replace('é','e')
```
